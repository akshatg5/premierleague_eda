# Exploratory Data Analysis Project
#### Analysing the English Premiere League season 2020-21. 



The English Premiere League is one of the most-watched football Leagues in the world with over 4 billion views every season. Established in 1992, the league consist of 20 teams that play each other twice, in home and away games over the year, with the team accumulating most points taking the trophy. 

Owing to the mass viewership EPL clubs enjot vast pool of financial resources. These huge bank banalances allow them to involve the best players around the world, coached by the best managers and state of the art training facilities. 

Data Analysis has become increasingly important with the involvement of technology in football,the clubs use advanced statistical models for recruitment, training and tactical decisions. Data Science is playing a huge role in the football industry, ranging from player's performance tracking to sentiment analysis on social media, technology has transformed the way teams compete and operate. 

As a fan of Manchester City , EPL has been great for me, credit to the Man City team for winning 6 out of the last 10 seasons. The EPL also features famous teams like Manchester United, Arsenal, Chelsea, Tottenham Hotspur and more. 

```python
#Importing Libraries 
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

## Dataset :

The Dataset used in the project is English Premier League 2020-2021 dataset from Kaggle. The Dataset consists of data about the EPL players in the season 2020-21. The Data has indivisual player data that includes number of goals scored, assists, starts, minutes played by the players.  

```python
epl_df_raw = pd.read_csv('EPL 2020_2021.csv')
epl_df_raw
```

#### Data Details

```python
epl_df_raw.info()
```

```python
epl_df_raw.describe()
```

## Data Preparation and Cleaning


```python
#counting the missing values - depicted by NaN wherever a value is missing. 
epl_df_raw.isna().sum()
```

Clearly, there are no missing values in the dataset. This depicts that the dataset is a very well compiled dataset where all values are present. Credit to the creator of the dataset.

Creating a few extra columns that will help in data visualization and overall data analysis.

```python
epl_df_raw['GoalsPerMatch'] = (epl_df_raw['Goals']/epl_df_raw['Matches']).astype(float)
epl_df_raw['MinutesPerMatch'] = (epl_df_raw['Mins']/epl_df_raw['Matches']).astype(int)
epl_df_raw
```

# Exploratory Analysis and Visualization

#### Players from different nationalities playing in the EPL in the season 2020-21

```python
epl_df_raw['Nationality'].unique()
```

It is clear that players from multiple nations take part in the EPL.

```python
nations = np.size(epl_df_raw['Nationality'].unique())
print(f"Players from {nations} nations take part in the English Premier League.")
```

```python
players_nation
```

```python
sns.set_style('whitegrid')
```

```python
fig,ax = plt.subplots(facecolor = 'white',figsize = (20,18))
labels = ['ENG', 'FRA', 'BRA', 'ESP', 'IRL', 'POR', 'SCO', 'NED', 'WAL', 'BEL',
       'GER', 'ARG', 'CIV', 'NGA', 'USA', 'DEN', 'SUI', 'SEN', 'EGY', 'ITA',
       'POL', 'SWE', 'GHA', 'COL', 'NIR','TUR']
data = players_nation.head(26)
colors = sns.color_palette('pastel')
ax.set_title("EPL Player's Nationality",fontsize = 20)
ax.pie(data,labels = labels,colors=colors,autopct='%1.1f%%');
```

The given pie chart makes it clear that after England, other european countries have the highed participation percentage. Majority of the participants belong to nations where football is very famous. 

#### Calculating number of goals and penalties scored in the season. 

```python
goals_scored = epl_df_raw['Goals'].sum()
print(goals_scored)
```

```python
penalty_scored = epl_df_raw['Penalty_Goals'].sum()
print(penalty_scored)
```

```python
penalty_attempts = epl_df_raw['Penalty_Attempted'].sum()
print(penalty_attempts)
```

#### Pie Chart - Penalties Scored VS Penalties Missed

```python
fig,ax = plt.subplots(facecolor='white',figsize=(8,8))
pen_missed = penalty_attempts - penalty_scored
data = [pen_missed,penalty_scored]
labels = ['Missed','Scored']
colors = ['#FF5733', '#32CD32'];
ax.pie(data,labels = labels,colors=colors,autopct='%1.1f%%');
ax.set_title('Penalties Scored Vs. Missed',fontsize=25);
```

## Players belonging to different age groups.

In the EPL, the players' age ranges from 16 to 38, where most of the players are 26 years old.

```python
df_ages = pd.DataFrame({'values' :[78, 186, 197,71]})
age_index = ['under 20','between 20-25','between 25-30','above 30']
df_ages = df_ages.set_index(pd.Index(age_index,name='Age Groups'))
df_ages
```

```python
fig,ax = plt.subplots(figsize=(15,10))
colors = sns.color_palette('tab10');
squarify.plot(sizes=df_ages['values'],label=df_ages.index,color=colors,text_kwargs={'fontsize':20});
ax.set_title('Age Groups of EPL season 20-21',fontsize=25);
```

Using this TreeMap we can conclude that maximum players belong to the age group 20 to 30 years. Only a small portion of the players participating arre either under 20 or above 30. Indicating that top performance of a footballer lies between age 20 to 30.

#### Average age of players in each club:

```python
fig,ax = plt.subplots(figsize = (22,10))
plt.grid(True,axis='y');
sns.boxplot(data=epl_df_raw,x='Club',y='Age',palette='dark', boxprops=dict(alpha=.9));
plt.xticks(rotation=90);
ax.set_title('Average Age of Players from each club',fontsize=20);
ax.set_xlabel('Club Name',fontsize=20);
ax.set_ylabel('Age',fontsize=20);
plt.xticks(fontsize=16);
plt.yticks(fontsize=16);
```

Using this Box plot we can conclude a few points: 

1. Manchester City has a team wherein the median age is around 25, the oldest player is 35 years old and youngest is under 20. The lines extending out of the box are approximately equal in length, which signifies that the age distribution is symmetric.
2. Manchester United's team has a median age of around 20, their oldest player is also 32-33 year old and the whole box plot lies closer to the 20s indicating their team is quite young.

Similarly using box plot's properties we can conclude a lot of points about each team and their age distribution. 

## Assists and Goals 

This part will cover the number of goals scored and assists provided by each club. 

#### Goals Scored by Each club

```python
goals_club = pd.DataFrame(epl_df_raw.groupby('Club',as_index=False)['Goals'].sum())
fig,ax = plt.subplots(figsize = (22,10))
sns.barplot(data=goals_club.sort_values(by='Goals'),x="Club",y='Goals')
for i, v in enumerate(goals_club['Goals'].sort_values()):
    ax.text(i, v + 1, str(v), ha='center', fontsize=14)
ax.set_title("Goals Scored By Each Club",fontsize=25)
plt.xticks(rotation=90,fontsize=18);
ax.set_ylabel("Goals",fontsize=20);
ax.set_xlabel("Club",fontsize=20);
```

#### Assists given by Each Club

```python
goals_club = pd.DataFrame(epl_df_raw.groupby('Club',as_index=False)['Assists'].sum())
fig,ax = plt.subplots(figsize = (22,10))
sns.barplot(data=goals_club.sort_values(by='Assists'),x="Club",y='Assists')
for i, v in enumerate(goals_club['Assists'].sort_values()):
    ax.text(i, v + 1, str(v), ha='center', fontsize=14)
ax.set_title("Assists given By Each Club",fontsize=25)
plt.xticks(rotation=90,fontsize=18);
ax.set_ylabel("Goals",fontsize=20);
ax.set_xlabel("Club",fontsize=20);
```

#### Highest Goals scorers of the 2020-21 season

The following section will include the best performers of the season, the top 10 players with most goals.

```python
df_scoring = epl_df_raw[['Name','Club','Matches','Assists','Goals']]
top_10_scorers = df_scoring.sort_values(by='Goals',ascending=False).nlargest(n=10,columns='Goals')
top_10_scorers
```

```python
fig,ax=plt.subplots(figsize=(24,12))
sns.scatterplot(data=top_10_scorers, x='Goals', y='Assists', hue='Name',s=100);
for i,row in top_10_scorers.iterrows():
    ax.annotate(row['Name'],xy=(row['Goals'],row['Assists']),xytext=(row['Goals']-0.2,row['Assists']+0.2),fontsize=18)
ax.legend(bbox_to_anchor=(1.05,1),loc = 'upper left',fontsize=20);
ax.set_title('Top 10 Goal Scorers',fontsize=25);
ax.set_xlabel('Goals',fontsize=20);
ax.set_ylabel('Assists',fontsize=20);
```

This scatter plot consists of the top 10 Goals scored in the season 2020-21.Harry Kane is at the top of the goal scoring list followed by Bruno Fernandes. The plot also show cases their number of goals scored and the assists they made over the season.Ikay Gundogan and Alexandre Lacazette have the same number of goals and assists and hence their names overlap. 

The following section will include the best performers of the season, the top 10 players with most goals.

```python
top_assists = epl_df_raw[['Name','Club','Matches','Assists','Goals']]
top_10_assists = top_assists.nlargest(n=10,columns='Assists')
```

```python
top_10_assists
```

```python
fig,ax = plt.subplots(figsize=(20,12))
sns.scatterplot(data=top_10_assists,x='Goals',y='Assists',hue='Name',s=100);
for i,row in top_10_assists.iterrows():
    ax.annotate(row['Name'],xy=(row['Goals'],row['Assists']),xytext=(row['Goals']-0.2,row['Assists']+0.2),fontsize=20)
ax.legend(bbox_to_anchor = (1.05,1),loc='upper left',fontsize=20);
ax.set_title('Top 10 Assists providers',fontsize=25);
ax.set_xlabel('Goals',fontsize=20);
ax.set_ylabel('Assists',fontsize=20);
```

The plot shows that Harry Kane gave the most assists and also scored most goals. Followed by Kevin and Bruno Fernandes at the second position.

#### Q.1 Which player played for the most minutes?

```python
most_min = epl_df_raw.sort_values(by='Mins',ascending=False)
top_10_mins = most_min.head(10)
top_10_mins
```

```python
fig,ax = plt.subplots(figsize=(20,12))
sns.barplot(data = top_10_mins,x="Name",y="Mins")
ax.set_title("Most Minutes of the match Played",fontsize=25);
ax.set_xlabel("Name of the Players",fontsize=20);
ax.set_ylabel("Number of minutes played",fontsize=20);
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right", fontsize=16);
```

```python
max_mins = epl_df_raw['Mins'].max()
print(f"Maximum minutes played : {max_mins}")
players_maxmin = epl_df_raw.loc[epl_df_raw['Mins']== 3420]
players = players_maxmin['Name'].values
players
names_str = ""
for name in players:
    names_str += name + ","
print(f"6 players played for {max_mins} minutes who are {names_str}.")
```

#### Q.2 Which player has the most goals and assists for the season?

```python
fig,ax = plt.subplots(figsize=(20,25))
sns.scatterplot(data=top_10_assists,x='Goals',y='Assists',hue='Name',s=100);
for i,row in top_10_assists.iterrows():
    ax.annotate(row['Name'],xy=(row['Goals'],row['Assists']),xytext=(row['Goals']-0.2,row['Assists']+0.2),fontsize=20)
ax.legend(bbox_to_anchor = (1.05,1),loc='upper left',fontsize=20);
ax.set_title('Top 10 Assists providers',fontsize=25);
ax.set_xlabel('Goals',fontsize=20);
ax.set_ylabel('Assists',fontsize=20);
```

```python
player = top_10_assists.nlargest(n=1,columns=['Assists','Goals'])['Name'].values[0]
print(f"The player with maximum number of goals and assists is {player}.")
```
